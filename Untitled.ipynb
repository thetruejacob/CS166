$$1=(1-p)\sum_{k=0}^{8} kBinom(8, p) + p\sum_{k=0}^{8} kBinom(8, p) = 1 * \sum_{k=0}^{8} kBinom(8, p) $$

    The sum of probabilities for all possible outcomes for an equal binomial will always be 1, thus 1=1.


```python
# Simple CA simulator in Python
#
# *** Hosts & Pathogens ***
#
# Copyright 2008-2012 Hiroki Sayama
# sayama@binghamton.edu

import matplotlib
matplotlib.use('TkAgg')

import pylab as PL
import random as RD
import scipy as SP

RD.seed()

width = 50
height = 50

def init():
    global time, config, nextConfig, ratio

    time = 0
    
    config = SP.zeros([height, width])
    for x in range(width):
        for y in range(height):
            if RD.random() < initProb:
                state = 2
            else:
                state = 1
            config[y, x] = state

    nextConfig = SP.zeros([height, width])
    ratio = []

def draw():
    PL.cla()
    PL.subplot(1,2,1)
    PL.pcolor(config, vmin = 0, vmax = 2, cmap = PL.cm.jet)
    PL.axis('image')
    PL.title('t = ' + str(time))

    ratio.append(sum([c == 1 for row in config for c in row])/sum([c == 2 for row in config for c in row]))
    PL.subplot(1,2,2)
    PL.title("Healthy / Infected cells")
    PL.plot(ratio)

def step():
    global time, config, nextConfig

    time += 1

    for x in range(width):
        for y in range(height):
            state = config[y, x]
            if state == 0:
                for dx in range(-1, 2):
                    for dy in range(-1, 2):
                        if config[(y+dy)%height, (x+dx)%width] == 1:
                            if RD.random() < regrowthRate:
                                state = 1
            elif state == 1:
                for dx in range(-1, 2):
                    for dy in range(-1, 2):
                        if config[(y+dy)%height, (x+dx)%width] == 2:
                            if RD.random() < infectionRate:
                                state = 2
            else:
                state = 0

            nextConfig[y, x] = state

    config, nextConfig = nextConfig, config

def setInitProb(val=.02):
	global initProb
	initProb = val
	return initProb

def setInfectionRate(val=.85):
	global infectionRate
	infectionRate = val
	return infectionRate

def setRegrowthRate(val=.15):
	global regrowthRate
	regrowthRate = val
	return regrowthRate

import pycxsimulator
pycxsimulator.GUI(parameterSetters= [setInitProb, setInfectionRate, setRegrowthRate]).start(func=[init,draw,step])
```
