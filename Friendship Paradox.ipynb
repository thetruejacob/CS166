### The Friendship Paradox
On average, your friends have more friends than you do. 

Average number of friends
$$ \mu = \frac{\sum_{v \in V} d(v)}{2|E|} = \frac{2 |E|}{|V|}$$

Average number of friends of typical friends
$$ \frac{\sum_{v \in V}d(v)^2}{2|E|} = \mu + \frac{\sigma^2}{\mu} > \mu$$

### Proof
Notation:
- $V$: The set of all vertices
- $n$: The number of vertices
- $v$: a single vertex
- $d(v)$: The degree of vertex v

The average number of friends is the sum of the degrees of each vertex, divided by the number of vertices. 
$$ \mu = \frac{\sum_{v \in V} d(v)}{n}$$

For the second part, consider a single vertex $v$. Each $v$ must ask each of its $d(v)$ friends how many friends they have, so each $v$ will obtain $d(v)$ lists. There are in total $\sum_{v \in V} d(v)$ lists. As for the total number of names on lists, consider that each $v$ has $d(v)$ friends, and each of the $d(v)$ friends will also have a record stating that $v$ has $d(v)$ friends. This is a list of length $d(v)$. So there are in total $d(v)^2$ names generated by $v$. Thus the average number of friends of friends is 
$$ \frac{\sum_{v \in V} d(v)^2}{\sum_{v \in V} d(v)} $$

The cauchy-schwarz inequality states that the dot product of two vectors is less than or equal to the dot product of their norms. Then consider two vectors, both of length $v$, one containing all 1s, and the other containing $d(v)$s.

$$ \begin{bmatrix} 1 \\ 1 \\ ... \end{bmatrix} \cdot \begin{bmatrix}d(v_1) \\ d(v_2) \\ ... \end{bmatrix} \leq || \begin{bmatrix} 1 \\ 1 \\ ... \end{bmatrix} || \cdot || \begin{bmatrix} d(v_1) \\ d(v_2) \\ ... \end{bmatrix} || $$
 
$$ \sqrt{d(v_1) + d(v_2)  + ...}\leq \sqrt{n} \cdot \sqrt{d(v_1)^2 + d(v_2)^2  + ...}$$

$$ \bigg(\sum_{v \in V}d_v \bigg)^2 \leq n \sum_{v \in V} d(v)^2$$
Rearranging,
$$ \mu =  \frac{\sum_{v \in V} d(v)}{n} \leq \frac{\sum_{v \in V} d(v)^2}{\sum_{v \in V} d(v)} $$
