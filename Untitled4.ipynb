```python
import math
import scipy.stats as ss


def fit_decay(t, theta):
    '''The fit function is the exponential function described in Eq. (1) of the Study Guide.
       theta = (N_0, lambda) = (theta[0], theta[1]) is the parameters vector.'''
    return theta[0]*np.exp(-theta[1] * t)

def log_likelihood(t_values_raw, count_values_raw, theta):
    '''Computes the natural logarithm of the likelihood function for a given set of parameters'''
    sum_least_squares = np.sum((fit_decay(t_values_raw, theta) - count_values_raw)**2)
    return -sum_least_squares/len(t_values_raw)

theta_guess = [1, 1] ## notice these are completely off-target!
current_log_likelihood = log_likelihood(t_values_raw, count_values_raw, theta_guess)

mcmc_results = [] ## let's store the proposals for the parameters in a list
n_trials = 10000  ## number of trials, or iterations, to be run
np.random.seed(seed=123) ## Setting the seed for reproducibility

for i in range(n_trials):
    ## propose new theta using random noise
    new_theta = [theta + 1*np.random.randn() for theta in theta_guess]
    
    ###################################### Fill in the blanks below

    new_log_likelihood = log_likelihood(data_raw['time'], data_raw['count_rate'], new_theta)

    ## likelihood ratio
    r = np.exp(new_log_likelihood - current_log_likelihood)

    ## criterion selection (do we accept the new proposal?)

    if r > np.random.random():
        current_log_likelihood = new_log_likelihood
        theta_guess = new_theta
    ## append the new results to mcmc_results
    mcmc_results.append(theta_guess)
```
